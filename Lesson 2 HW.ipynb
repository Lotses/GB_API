```python
from pprint import pprint
from lxml import html
import requests
```

```python
# адрес для парсинга новостей
url = 'https://lenta.ru'
# заголовки
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

# делаем запрос
response = requests.get(url, headers=headers)
```

```python
# проверяем коннект
response.ok
```

```python
# переводим ответ в дом
dom = html.fromstring(response.text)
```

```python
# Парсим названия статей
text_list = dom.xpath("//div/span/text()")
text_list
```

```python
# Парсим время
date_list = dom.xpath("//div/time/text()")
date_list
```

```python
# Парсим ссылки
ref_list = dom.xpath("//div[contains(@class, 'card-mini__text')]/../@href")
ref_list
```
